1. Which 5 restaurants have the highest income?

```python
select
	name
	, revenue
from restaurant_data rd
group by 1, 2
order by 2 desc
limit 5;
```

2. What is the average rating of restaurants in each location?

```python
select
	Location
	, round(avg(rating), 4) as rating_location
from restaurant_data rd
group by 1
order by 2 desc
;
```

2.png

3. What types of kitchens bring the highest average income?

```python
select
	cuisine
	, round(avg(revenue), 0) as avg_revenue
from restaurant_data rd
group by 1
order by 2 desc
limit 1
;
```

3.png

4. How does a chef's experience affect a restaurant's rating?

```python
select
	case
		when "Chef Experience Years" < 6 then '0-5 years'
		when "Chef Experience Years" between 6 and 10 then '6-10 years'
		else 'over 10 years' end as exp_group
	, round(avg(rating), 4) as avg_rating
from restaurant_data rd
group by exp_group
;
```

4.png

5. Which restaurants have the best ratio of revenue to marketing budget?

```python
select
	name
	, revenue
	, `Marketing Budget`
	, round((revenue - `Marketing Budget`) / `Marketing Budget`, 1) as ROMI
from restaurant_data
group by 1,2,3
order by 4 desc
limit 5
;
```

7. Is there a correlation between the number of seats and restaurant revenue?

```python
select
	corr(`Seating Capacity`, Revenue)  as correlation
from restaurant_data
;
```

7.png

The correlation between the number of seats in a restaurant and revenue, which is 0.67, indicates a positive relationship.
Positive trend: Since the correlation is close to 1, an increase in the number of seats in a restaurant is likely to increase revenue.
This can mean that the more guests a restaurant can accommodate, the greater the potential for revenue growth.Помірний зв'язок: Хоча кореляція досить висока, вона все ж не є ідеальною.
This shows that other factors besides the number of seats also play an important role in revenue generation.

8. How does the average price of a dish differ in restaurants with different types of cuisine?

```python
SELECT
  cuisine
  , round(avg(`Average Meal Price`), 1) as avg_price
FROM restaurant_data
group by 1
order by 2 desc
LIMIT 5
;
```

9. Which restaurants have the biggest difference between the number of bookings on weekends and weekdays?

```python
SELECT
  Name
  , (`Weekend Reservations` - `Weekday Reservations`) as diff_day1
FROM restaurant_data
group by 1, `Weekend Reservations`, `Weekday Reservations`
order by 2 desc
LIMIT 3
;
```
